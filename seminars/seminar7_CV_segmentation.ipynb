## Segmentation

### Какие метрики использовать для сегментации?

#### 1. Pixel accuracy

Не стоит использовать, т.к. дезинформирует при дисбалансе классов. А в сегментации несбалансированность встречается не реже и, скорее, даже чаще, чем в задаче классификации

#### 2. Dice coefficient

Эквивалентен F1 score

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/2/26/Precisionrecall.svg/350px-Precisionrecall.svg.png" alt="precision_recall" width=300px/>

$$ \Large IoU = \dfrac{2TP}{(TP + FP) + (TP + FN)} $$


#### 3. IoU - intersection over union или Jaccard coefficient

Мера того, насколько хорошо ground truth пересекается с предсказаниями модели

<img src="https://www.pyimagesearch.com/wp-content/uploads/2016/09/iou_equation.png" alt="IoU" width=300px/>

$$ \Large IoU = \dfrac{TP}{TP + FP + FN} $$


#### 4. mean average precision at different IoU

Cчитается путем усреднения AP для разных порогов IoU, обычно `range(0.5, 1.0, 0.05)`. Такой подход поощряет модели, которые дают лучшую локализацию объекта

Пример использования: [Kaggle TGS Salt Identification](https://www.kaggle.com/c/tgs-salt-identification-challenge#evaluation)

```
The metric sweeps over a range of IoU thresholds, at each point calculating an average precision value. The threshold values range from 0.5 to 0.95 with a step size of 0.05: (0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95).

In other words, at a threshold of 0.5, a predicted object is considered a "hit" if its intersection over union with a ground truth object is greater than 0.5.

At each threshold value t, a precision value is calculated based on the number of true positives (TP), false negatives (FN), and false positives (FP) resulting from comparing the predicted object to all ground truth objects.

A true positive is counted when a single predicted object matches a ground truth object with an IoU above the threshold.
A false positive indicates a predicted object had no associated ground truth object.
A false negative indicates a ground truth object had no associated predicted object.

The average precision of a single image is then calculated as the mean of the above precision values at each IoU threshold.
```


#### 5. mean F2 score at different IoU

Считается усреднением F2 для разных порогов IoU

Пример использования: [Kaggle Airbus Ship Detection](https://www.kaggle.com/c/airbus-ship-detection#evaluation)

### А какой loss?

#### 1. loss для классификации

Т.к. сегментация - это, по сути, классификация пикселей, то можем использовать loss для этой задачи, например, кросс-энтропию или [focal loss](https://arxiv.org/abs/1708.02002) (подробнее в ноутбуке про object detection)

Минус: никак не учитывается локализация объекта на изображении

#### 2. Попытаться оптимизировать dice или jaccard

Явно сделать это не получится (функции недифференцируемы), поэтому используются [soft dice](http://campar.in.tum.de/pub/milletari2016Vnet/milletari2016Vnet.pdf) и soft jaccard - варианты функций, когда вместо дискретизации по порогу и подсчета TP, FP, TN, FP, используются вероятности принадлежности к определенному классу

<img src="https://www.jeremyjordan.me/content/images/2018/05/Screen-Shot-2018-05-24-at-10.50.59-PM.png" alt="IoU" width=800px/>

На практике используют $ 1 - Dice $, чтобы минимизировать функцию потерь

#### 3. Смешать пункты 1 и 2

Варианты:

- $CE + (1 - Dice)$
- $0.5 * CE + 0.5 * (1 - Dice)$
- $\alpha * CE + (1 - \alpha) * (1 - Dice)$


#### 4. Lovasz loss

**!!! НЕ Лаваш или Ловаш, а Ловас**

В задаче сегментации частой проблемой являются предсказания границ объектов. 

Одним из решений является перевзвешивание лосса для пикселей границ объектов. Lovasz loss - другое решение данной проблемы

- Цель lovasz loss - уточнить предсказания на границах объектов
- Стоит использовать только для файнтюнинга весов, обучать с нуля не получится
- [paper](https://arxiv.org/abs/1705.08790)
- [pytorch and TF code](https://github.com/bermanmaxim/LovaszSoftmax)


Реализуем dice

````python
import torch


def soft_dice(*, y_true, y_pred):
    eps = 1e-15
    y_pred = y_pred.contiguous().view(y_pred.numel())
    y_true = y_true.contiguous().view(y_true.numel())
    intersection = (y_pred * y_true).sum(0)
    scores = 2. * (intersection + eps) / (y_pred.sum(0) + y_true.sum(0) + eps)
    score = scores.sum() / scores.numel()
    return torch.clamp(score, 0., 1.)


def hard_dice(*, y_true, y_pred):
    y_pred = torch.round(y_pred)
    return soft_dice(y_true=y_true, y_pred=y_pred)
````

````python
from torch import nn
from torch.nn.modules.loss import BCEWithLogitsLoss

class DiceLoss(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, inputs, target):
        return 1 - soft_dice(y_true=target, y_pred=torch.sigmoid(inputs))


class BCEDiceLoss(nn.Module):
    def __init__(self, dice_weight=0.5):
        super().__init__()
        self._dice = DiceLoss()
        self._dice_weight = dice_weight

    def forward(self, inputs, target):
        return (1 - self._dice_weight) * BCEWithLogitsLoss()(inputs, target) + \
            self._dice_weight * self._dice(inputs, target)
````

### UNet

[paper](https://arxiv.org/abs/1505.04597)


<img src="https://lmb.informatik.uni-freiburg.de/people/ronneber/u-net/u-net-architecture.png" alt="unet" width=800px/>

````python
import torch
from torch import nn


class ConvRelu(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, 3, padding=1)
        self.activation = nn.ReLU(inplace=True)

    def forward(self, x):
        x = self.conv(x)
        x = self.activation(x)
        return x

    
class DoubleConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv_block = nn.Sequential(
            ConvRelu(in_channels, out_channels),
            ConvRelu(out_channels, out_channels),
        )
    
    def forward(self, x):
        x = self.conv_block(x)
        return x


class EncoderBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(EncoderBlock, self).__init__()
        self.conv_block = DoubleConvBlock(in_channels, out_channels)
        self.max_pool = nn.MaxPool2d(kernel_size=2, stride=2)
            
    def forward(self, x):
        before_pool = self.conv_block(x)
        x = self.max_pool(before_pool)
        return x, before_pool


class DecoderBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DecoderBlock, self).__init__()              
        self.conv_block = DoubleConvBlock(in_channels, out_channels)

    def forward(self, x, y):
        x = nn.functional.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        return self.conv_block(torch.cat([x, y], dim=1))


class UNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1, n_filters=64):
        super().__init__()
        self.pool = nn.MaxPool2d(2, 2)

        self.enc1 = EncoderBlock(in_channels, n_filters)
        self.enc2 = EncoderBlock(n_filters, n_filters * 2)
        self.enc3 = EncoderBlock(n_filters * 2, n_filters * 4)
        self.enc4 = EncoderBlock(n_filters * 4, n_filters * 8)
        
        self.center = DoubleConvBlock(n_filters * 8, n_filters * 16)
        
        self.dec4 = DecoderBlock(n_filters * (16 + 8), n_filters * 8)
        self.dec3 = DecoderBlock(n_filters * (8 + 4), n_filters * 4)
        self.dec2 = DecoderBlock(n_filters * (4 + 2), n_filters * 2)
        self.dec1 = DecoderBlock(n_filters * (2 + 1), n_filters)

        self.final = nn.Conv2d(n_filters, out_channels, kernel_size=1)

    def forward(self, x):
#         print('x', x.shape)
        x, enc1 = self.enc1(x)
#         print('enc1', enc1.shape, 'x', x.shape)
        x, enc2 = self.enc2(x)
#         print('enc2', enc2.shape, 'x', x.shape)
        x, enc3 = self.enc3(x)
#         print('enc3', enc3.shape, 'x', x.shape)
        x, enc4 = self.enc4(x)
#         print('enc4', enc4.shape, 'x', x.shape)

        center = self.center(x)
#         print('center', center.shape)

        dec4 = self.dec4(center, enc4)
#         print('dec4', dec4.shape)
        dec3 = self.dec3(dec4, enc3)
#         print('dec3', dec3.shape)
        dec2 = self.dec2(dec3, enc2)
#         print('dec2', dec2.shape)
        dec1 = self.dec1(dec2, enc1)
#         print('dec1', dec1.shape)
        
        final = self.final(dec1)
#         print('final', final.shape)
        return final
````

Посчитаем mean и std для пикслей изображений в train

````python
train_images = np.array([sample[0][0] / 255.0 for sample in train_data])

TRAIN_MEAN = np.mean(train_images)
TRAIN_STD = np.std(train_images)

print(TRAIN_MEAN, TRAIN_STD)
````

````python
from albumentations import Compose, OneOf, ShiftScaleRotate, RandomCrop, RandomRotate90, \
    Flip, RandomContrast, RandomBrightness, Normalize
from torchvision.transforms.functional import to_tensor


class Transformer:
    def __init__(self):
        pass
    
    def transform(self, img, mask):
        raise NotImplementedError
    

class ValTransformer(Transformer):
    def __init__(self):
        super().__init__()
        self._normalize = Normalize(mean=(TRAIN_MEAN), std=(TRAIN_STD), p=1.0)

    def transform(self, img, mask):
        img = self._normalize.apply(image=img)
        mask = (mask > 0).astype(np.uint8) * 255
        return to_tensor(img), to_tensor(mask)
        

class TrainTransformer(Transformer):
    def __init__(self):
        super().__init__()
        self._augmentation = Compose([
            RandomCrop(256, 256, p=1.0),
            RandomRotate90(),
            Flip(),
            ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.1, rotate_limit=15, p=0.5),
            OneOf([
                RandomContrast(p=1.0),
                RandomBrightness(p=1.0)
            ], p=0.3),
            Normalize(mean=(TRAIN_MEAN), std=(TRAIN_STD), p=1.0)
        ])

    def transform(self, img, mask):
        data = {'image': img, 'mask': mask}
        augmented = self._augmentation(**data)
        augmented_img = augmented['image']
        augmented_img = to_tensor(augmented_img)
        
        augmented_mask = augmented['mask']
        augmented_mask = (augmented_mask > 0).astype(np.uint8) * 255
        augmented_mask = to_tensor(augmented_mask)
        
        return augmented_img, augmented_mask
````

Теперь можно реализовать и сам класс датасета

````python
import os

import cv2
from tqdm import tqdm
from scipy.misc import imread, imresize
from torch.utils.data import Dataset, DataLoader

        
class ISBIDataset(Dataset):
    def __init__(self, data, transformer):
        self._data = data
        self._transformer = transformer
        
    def __getitem__(self, idx):
        img, mask = self._data[idx]
        img = np.expand_dims(img, axis=2)
        mask = np.expand_dims(mask, axis=2)
        img, mask = self._transformer.transform(img, mask)
        return {'img': img, 'mask': mask}

    def __len__(self):
        return len(self._data)
````

````python
batch_size = 4
val_batch_size = 1 # будем на валидации считать метрики по одному изображению
num_workers = 0 # Windows multiprocessing bug


train_transformer = TrainTransformer()
train_dataset = ISBIDataset(train_data, train_transformer)
train_loader = DataLoader(train_dataset, batch_size, shuffle=True, num_workers=num_workers)

val_transformer = ValTransformer()
val_dataset = ISBIDataset(val_data, val_transformer)
val_loader = DataLoader(val_dataset, val_batch_size, shuffle=False, num_workers=num_workers)
````

`Trainer` воозьмем из предыдущего семинара

````python
import os
from collections import defaultdict

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.optim import Adam
from tqdm import tqdm


class Trainer:
    CHECKPOINTS_PATH = 'checkpoints'
    
    def __init__(self, model, criterion, metrics, config, device='cuda'):
        self._model = model
        self._device = device
        self._criterion = criterion
        self._metrics = metrics
        
        self._model.to(self._device)
        
        self._epochs = config['epochs']             
        self._early_stopping = config['early_stopping']
        self._optimizer = Adam(self._model.parameters(), lr=config['lr'])
        self._scheduler = ReduceLROnPlateau(self._optimizer,
                                            'min',
                                            factor=config['lr_reduce_rate'],
                                            patience=config['patience'],
                                            verbose=True)

        
        self._best_loss = float('inf')
        self._best_epoch = -1
        self._best_epoch_dice = float('-inf')

        self._history = {'train': defaultdict(list), 'val': defaultdict(list)}
        
        if not os.path.exists(Trainer.CHECKPOINTS_PATH):
            os.makedirs(Trainer.CHECKPOINTS_PATH)

    @property
    def history(self):
        return self._history
        
    def fit(self, train_loader, val_loader):
        for epoch in range(self._epochs):
            self._model.train()
            train_metrics = self._run_epoch(epoch, train_loader, is_training=True)
            print('Train: ' + ' | '.join(self._get_metrics_string(train_metrics, is_training=True)))
                       
            self._model.eval()
            val_metrics = self._run_epoch(epoch, val_loader, is_training=False)
            print('Validation: ' + ' | '.join(self._get_metrics_string(val_metrics, is_training=False)))
            
            val_loss = val_metrics['loss']
            self._scheduler.step(val_loss)
           
            if self._best_loss > val_loss:
                self._best_loss = val_loss
                self._best_epoch = epoch
                self._best_epoch_dice = val_metrics['hard_dice']
                self._save_model(epoch)
            
            self.plot_history()
            
            if abs(self._best_epoch - epoch) > self._early_stopping:
                print('Early stopping. Epoch {}. Best epoch: {}, best loss: {}'.
                      format(epoch, self._best_epoch, self._best_loss))
                break
        
    def _run_epoch(self, epoch, loader, is_training):
        if is_training:
            pbar = tqdm(enumerate(loader), total=len(loader), desc='Epoch {}'.format(epoch))
        else:
            pbar = enumerate(loader)
        
        avg_metrics = defaultdict(float)
        for i, data in pbar:
            batch_metrics = self._step(data, is_training)
            for name, val in batch_metrics.items():
                avg_metrics[name] += val
        
        return {name: value / len(loader) for name, value in avg_metrics.items()}
    
    def _step(self, data, is_training=True):
        metrics_values = {}
        images = data['img'].to(self._device)
        y_true = data['mask'].to(self._device)

        if is_training:
            self._optimizer.zero_grad()

        with torch.set_grad_enabled(is_training):
            y_pred = self._model(images)
            loss = self._criterion(y_pred, y_true)
            
            for name, func in self._metrics:
                value = func(y_true=y_true, y_pred=torch.sigmoid(y_pred))
                metrics_values[name] = value.item()
            
            if is_training:
                loss.backward()
                self._optimizer.step()
        
        metrics_values['loss'] = loss.item()
        return metrics_values
    
    def _save_model(self, epoch):
        state = {
            'epoch': epoch,
            'state_dict': self._model.state_dict(),
            'best_loss': self._best_loss,
            'optimizer': self._optimizer.state_dict(),
        }
        fname = '{:.6f}_{:.6f}_{}.pt'.format(self._best_loss, self._best_epoch_dice, epoch)
        torch.save(state, os.path.join(Trainer.CHECKPOINTS_PATH, fname))
    
    def _get_metrics_string(self, metric_values, is_training):
        mode = 'train' if is_training else 'val'
        metrics_str = []
        for name, value in metric_values.items():
            self._history[mode][name].append(float(value))
            metrics_str.append('{}: {:.5f}'.format(name, float(value)))
        return metrics_str
    
    def plot_history(self):
        f = plt.figure(figsize=(12, 4))
        f.add_subplot(1, 2, 1)

        plt.plot(self.history['val']['hard_dice'], color='blue')
        plt.plot(self.history['train']['hard_dice'], color='green')
        plt.legend(['val', 'train'])
        plt.xlabel('epoch')
        plt.xticks(range(config['epochs']))
        plt.ylabel('dice')

        f.add_subplot(1, 2, 2)
        
        plt.plot(self.history['val']['loss'], color='blue')
        plt.plot(self.history['train']['loss'], color='green')
        plt.legend(['val', 'train'])
        plt.xlabel('epoch')
        plt.ylabel('loss')
        plt.xticks(range(config['epochs']))
        plt.show()
````

````python
import os
from collections import defaultdict

import numpy as np
import torch
import torch.nn as nn
from tqdm import tqdm


class Evaluator:    
    def __init__(self, model, metrics, device='cuda'):
        self._model = model
        self._metrics = metrics
        self._device = device

    def predict(self, loader, thr=0.5):
        y_pred = []
        self._model.eval().to(self._device)
        
        dices = []
        for i, data in tqdm(enumerate(loader), total=len(loader)):
            batch_pred, batch_dice = self._step(data, thr)
            y_pred.append(batch_pred)
            dices.append(batch_dice)

        dice = np.mean(dices)
        return y_pred, dice
    
    def _step(self, data, thr):
        images = data['img'].to(self._device)
        y_true = data['mask'].to(self._device)
        
        with torch.no_grad():
            y_pred = self._model(images)
               
        y_pred = y_pred.squeeze()
        y_pred = (torch.sigmoid(y_pred) > thr).type(y_pred.dtype)
        
        dice = hard_dice(y_true=y_true, y_pred=y_pred)
        
        return y_pred.cpu().numpy(), dice
````
