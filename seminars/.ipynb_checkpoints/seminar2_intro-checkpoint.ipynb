# Терминология ML

## Labels

Label (метка) - это то, что мы предсказываем. Это может быть цена дома, тип животного на картинке, часть речи слова - в общем, почти всё что угодно.

##Features

Feature (признак) - это входная переменная. Самая простая модель может использовать один единственный признак; более сложные модели могут использовать миллионы признаков.

Например, в задаче определения спам/не спам признаками могут быть:
* слова в тексте письма;
* адрес отправителя;
* время дня, в которое было отправлено письмо;
* наличие в письме фразы "Ваш адрес был выбран победителем";
* и др.

##Samples

Sample - это конкретный пример данных. Samples бывают двух видов:
* labeled (размеченные);
* unlabeled (неразмеченные).

Размеченный sample включает и признак, и label:

`  labeled examples: {features, label}: (x, y)`

Размеченные данные используются при обучении модели. Например, в задаче определения спам/не спам размеченными samples могут быть отдельные письма пользователей, которые явным образом были помечены как "спам" или "не спам".

Неразмеченный sample содержит только признак:

`  unlabeled examples: {features, ?}: (x, ?)`

После того, как мы обучили модель на размеченных данных, мы можем использовать модель, чтобы предсказывать labels для неразмеченных примеров. В определении спама неразмеченные примеры - это новые письма, ещё не размеченные людьми.

##Модели

Модели определяют зависимость между входными признаками и label. Например, модель, определяющая спам, может некоторые признаки чётко связывать с категорией "спам".

У модели можно выделить две фазы:
* Обучение модели: мы показываем модели размеченные примеры и даём возможность ей выучить связь между признаками и labels;
* Inference: мы применяем модель к неразмеченным примерам. То есть, мы используем модель, чтобы предсказать label (`y'`).



---

Итак, предположим, что Вы хотите разработать модель, которая будет предсказывать, является ли письмо спамом или нет. В качестве данных у нас есть набор писем, которые были помечены пользователями как "спам" или "не спам". Какие утверждения верны?


1.   Слова в теме письма будут хорошими labels.
2.   Вы будете использовать неразмеченные данные, чтобы обучить модель.
3. Те письма, которые не помечены "спам" или "не спам", - неразмеченные примеры.
4. Не все labels, которыми помечены письма, могут быть надёжными.

Теперь представим, что обувной он-лайн магазин хочет создать модель, которая будет рекомендовать пользователям обувь. То есть, модель будет рекомендовать определённые пары обуви Кейт и совершенно другие - Джону. Какие утверждения верны?
1. "Пользователь кликнул на описание обуви" - хороший label.
2. "Размер обуви" - полезный признак.
3. "Красота обуви" - полезный признак.
4. "Обувь, которую обожает пользователь" - хороший label.

#Train/val/test splitting

Целью модель машинного обучения является предсказывать labels для новых, ранее не виденных данных. Но если вы обучаете модель на Вашем наборе данных, то как Вы можете получить не виденные ранее данные? Один из возможных способов - разделить все данные на три части:
* обучающий набор - на котором учится модель;
* валидационный набор - для тестирования модели во время обучения;
* тестовый набор - данные для оценки конечной модели.

Разделение данных на три части (обучающую, валидационную и тестовую) очень важно при обучении модели.

Теперь установим и импортируем необходимые модули:

```python
import torch                # фреймворк автоматического дифференцирования
import torch.nn as nn       # модуль с основными блоками из чего состоят нейросети
import torch.optim as optim # модуль с оптимизаторами, как стохастический градиентный спуск и другие его модификации

from torch.utils.data import DataLoader # Класс для удобного итерирования по датасету
from torch.utils.data.dataset import Dataset # Обёртка над нашим датасетом

from torchvision.transforms import Resize # изменение размера картинки (матрицы) 
```

Создадим класс, который загружает данные:

```python
class CatsvsDogsData(Dataset):
    def __init__(self, folder_dataset, transform=None):
        self.__xs = []
        self.__ys = []
        self.transform = transform
        for f in os.listdir(folder_dataset):
            if '.DS_Store' in f:
                continue
            self.__xs.append(os.path.join(folder_dataset,f))
            if 'cat' in f:
                self.__ys.append(0)
            elif 'dog' in f:
                self.__ys.append(1)

    def __getitem__(self, index):

        img = Image.open(self.__xs[index])
        if self.transform is not None:
            img = self.transform(img)

        
        img = torch.from_numpy(np.asarray(img).astype(np.float32)) / 255.0
        img = img.reshape((3, 256, 256))
        label = self.__ys[index]
        return img, label
    
    def __len__(self):
        return len(self.__xs)
```

Загрузим данные:

```python
resize = Resize((256, 256))

dset_train = CatsvsDogsData(train_dir, transform=resize)
train_loader = DataLoader(dset_train, batch_size=100, shuffle=True, num_workers=1)

dset_val = CatsvsDogsData(val_dir, transform=resize)
val_loader = DataLoader(dset_val, batch_size=25, shuffle=True, num_workers=1)

dset_test = CatsvsDogsData(test_dir, transform=resize)
test_loader = DataLoader(dset_test, batch_size=25, shuffle=True, num_workers=1)
```

Так выглядит часть батча:

```python
imgs, targets = next(iter(train_loader))
print('Batch shape:',imgs.shape)
for i, image in enumerate(imgs):
    image = image.reshape((256, 256, 3))
    plt.imshow(image.numpy())
    plt.show()
    print('Class:', int(targets[i]))
    if i == 10:
        break
```

Зададим архитектруру нейронной сети:

```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 10, kernel_size=5),
            nn.MaxPool2d(kernel_size=2),
            nn.ELU(inplace=True)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(10, 20, kernel_size=5),
            nn.BatchNorm2d(20),
            nn.MaxPool2d(kernel_size=2),
            nn.ELU(inplace=True)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(20, 30, kernel_size=5),
            nn.BatchNorm2d(30),
            nn.MaxPool2d(kernel_size=2),
            nn.ELU(inplace=True)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(30, 30, kernel_size=5, stride=2),
            nn.BatchNorm2d(30),
            nn.MaxPool2d(kernel_size=2),
            nn.ELU(inplace=True)
        )
        
        self.feature_extractor = nn.Sequential(
            self.conv1,
            self.conv2,
            self.conv3,
            self.conv4
        )
        
        self.classifier = nn.Sequential(
            nn.Linear(1080, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(256, 2)
        )
    
    def forward(self, x):
        # x: [c, w, h]
        x = self.feature_extractor(x)        

        # global pooling is better
        x = x.view(-1, 1080)
        
        x = self.classifier(x)
        return x
```

Заведём объект, который отвечает за то, где будут лежать матрицы, на цпу или гпу.

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

Создадим саму сеть и определим функцию ошибки и оптимизатор:

```python
model = ConvNet().to(device)  # метод `to` позволяет "вручную" перенести матрицы на девайс (цпу или гпу)

criterion = nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)  # любимый оптимайзер
```

```python
def test(data_loader):
    with torch.no_grad():
        model.eval()
        test_loss = correct = 0
        for batch_idx, (data, target) in enumerate(data_loader):
            data, target = data.to(device), target.to(device)
            logits = model(data)
            loss = criterion(logits, target)
            
            test_loss += loss.item()
            pred = logits.max(1)[1]
            correct += pred.eq(target.view_as(pred)).cpu().sum()

        test_loss /= len(data_loader)
        acc = correct.numpy() / len(data_loader.dataset)
        print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({})\n'.format(
            test_loss,
            correct,
            len(data_loader.dataset),
            acc))
        
        return test_loss, acc
```

Теперь запустим обучение:

```python
num_epochs = 5
train_loss, train_acc = [], []
val_loss, val_acc = [], []


def train(epoch):
    model.train()
    losses, accs = [], []
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        logits = model(data)
        loss = criterion(logits, target)
        losses.append(loss.item())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
            epoch, batch_idx * len(data),
            len(train_loader.dataset),
            100 * batch_idx / len(train_loader),
            loss.item())
         )
    epoch_loss = np.sum(np.array(losses)) / len(losses)
    print('Train loss: {} '.format(epoch_loss))
    train_loss.append(epoch_loss)
        
        
for i in range(num_epochs):
    train(i)
    v_loss, v_acc = test(val_loader)
    val_loss.append(v_loss)
    val_acc.append(v_acc)
```

Построим графики со значениями функции ошибки и точности на протяжении обучения:

```python
plt.xlabel('Epoch #');
plt.ylabel('Accuracy');
plt.plot(val_acc);
plt.show();
```

Вычислим среднее значение ошибки и точность на тестовых данных:

```python
test(test_loader)
```

Посмотрим на предсказания для нескольких изображений:

```python
test_batch, targets = next(iter(test_loader))
for i, image in enumerate(test_batch[:20]):
    data = image.reshape((1, 3, 256, 256))
    data, target = data.to(device), targets[i].to(device)
    logits = model(data)
    pred = logits.max(1)[1]
    print("Predicted:", pred, "Target:",target)
    image = image.reshape((256, 256, 3))
    plt.imshow(image.numpy())
    plt.show()
```
